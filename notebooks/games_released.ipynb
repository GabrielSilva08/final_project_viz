```python
import pandas as pd
```

```python
df = pd.read_csv("games_2.csv")
```

```python
# 1. Converta para número (float), transformando valores inválidos em NaN
df["ReleaseYear"] = pd.to_numeric(df["ReleaseYear"], errors="coerce")

# 2. Remova linhas com NaN em ReleaseYear
df = df.dropna(subset=["ReleaseYear"])

# 3. Converta para inteiro
df["ReleaseYear"] = df["ReleaseYear"].astype(int)
```

```python
contagem = df["ReleaseYear"].value_counts()
```

```python
contagem = contagem[contagem.index <= 2025]
```

```python
contagem
```

```python
contagem_df = contagem.reset_index()
contagem_df.columns = ["ReleaseYear", "quantidade"]
```

```python
contagem_df
```

```python
contagem_df.to_json("contagem_por_ano.json", index=False)
```
